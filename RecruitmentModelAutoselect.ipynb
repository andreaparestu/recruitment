```python
## BINARY CLASSIFICATION PROBLEM: RECRUITMENT

# Compare Algorithms
from time import time
from pandas import read_csv
from matplotlib import pyplot
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.pipeline import Pipeline
from sklearn.pipeline import FeatureUnion
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest

# load dataset
filename = 'oldRecruitment.csv'
dataframe = read_csv(filename)
names = ['jenis_kelamin', 'umur', 'asal', 'penempatan', 'gaji', 'pendidikan', 'IPK', 'pengalaman','jenis_rekrut','id']
array = dataframe.values
X = dataframe[names]
Y = dataframe[['status_rekrut']]

# create feature union
features = []
features.append(('pca', PCA(n_components=3)))
features.append(('select_best', SelectKBest(k=6)))
feature_union = FeatureUnion(features)

feature_union.fit(X, Y)
mat = feature_union.transform(X)

# prepare models
models = []
models.append(('LR', LogisticRegression()))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))

# evaluate each model in turn
results = []
names = []
message = []
scoring = 'accuracy'
cur_score = 0
best_model = []
trained_models = []
for i, (name, model) in enumerate(models):
    kfold = KFold(n_splits=10, random_state=7)
    t0 = time()
    cv_results = cross_val_score(model, mat, Y, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(names)
    print("Process in train {} model finish in {:.3f}s".format(name, time() - t0))
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    if cv_results.mean() > cur_score:
        cur_score = cv_results.mean()
        best_model = i
        
    message.append(msg)
  
# Print Model Accuracy and SD
print(message)
print("{}: {}".format(best_model, cur_score))
clf = models[i][-1].fit(X, Y)
```
